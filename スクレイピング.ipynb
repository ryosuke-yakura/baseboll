```python
#2022年打席数
url = 'https://baseball-data.com/stats/hitter2-all/tpa-1.html'
#2021年打席数
url_2 = 'https://baseball-data.com/21/stats/hitter2-all/tpa-1.html'
#2020年打席数
url_3 = "https://baseball-data.com/20/stats/hitter2-all/tpa-1.html"
#2019年打席数
url_4 = "https://baseball-data.com/19/stats/hitter2-all/tpa-1.html"
#2018年打席数
url_5 = "https://baseball-data.com/18/stats/hitter2-all/tpa-1.html"
#2022年投手成績
url_6 = "https://baseball-data.com//stats/pitcher-all/"
#2021年投手成績
url_7 = "https://baseball-data.com/21//stats/pitcher-all/"
#2020年投手成績
url_8 = "https://baseball-data.com/20//stats/pitcher-all/"
#2019年投手成績
url_9 = "https://baseball-data.com/19//stats/pitcher-all/"
#2018年投手成績
url_10 = "https://baseball-data.com/18//stats/pitcher-all/"
#2022年年棒
url_11 = "https://baseball-data.com/ranking-salary/all/"
#2021年年棒
url_12 = "https://baseball-data.com/21/ranking-salary/all/"
#2020年年棒
url_13 = "https://baseball-data.com/20/ranking-salary/all/"
#2019年年棒
url_14 = "https://baseball-data.com/19/ranking-salary/all/"
#2018年年棒
url_15 = "https://baseball-data.com/18/ranking-salary/all/"
```

```python
html = urllib.request.urlopen(url_15)
soup = BeautifulSoup(html, "html.parser")
```

```python
#HTMLからtableの部分を全て取得
table = soup.find_all("table")[0]
#テーブルから行データを取得
rows = table.find_all("tr")
```

```python
#データ格納用
headData = []
playerData = []
```

```python
for i,row in enumerate(rows):    
    #1行目
  if i == 0:

      #ヘッダ部取得
      for headerValue in row.find_all('th'):
            headData.append(headerValue.get_text())
    #2行目～最終行未満
  elif (i > 0) and (i + 1 < len(rows)):
        #明細部取得
      playerRow = []
      for playerValue in row.find_all('td'):
          playerRow.append(playerValue.get_text())
      playerData.append(playerRow)
```

```python
#DataFrameに変換
df = pd.DataFrame(data = playerData, columns = headData)
```

```python
#CSV出力
df.to_csv('playerData_n_2018.csv', sep = ',', header = True, index = False)
```
